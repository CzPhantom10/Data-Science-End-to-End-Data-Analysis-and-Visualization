Part 2 : DATA WRANGLING & CLEANING

```python
# =========================================
# PART 2: DATA WRANGLING & CLEANING
# =========================================

import pandas as pd

# Load Dataset
df = pd.read_csv("smart_home_dataset.csv")

print("Dataset BEFORE Cleaning")
display(df.head())


# =========================================
# 1. Handling Missing Values
# =========================================
print("\nMissing Values BEFORE")
print(df.isnull().sum())

# Replace '?' with NaN
df.replace("?", pd.NA, inplace=True)

# Drop missing values
df = df.dropna()

print("\nMissing Values AFTER")
print(df.isnull().sum())


# =========================================
# 2. Removing Duplicates
# =========================================
print("\nDuplicates BEFORE:",
      df.duplicated().sum())

df = df.drop_duplicates()

print("Duplicates AFTER:",
      df.duplicated().sum())


# =========================================
# 3. Data Type Conversion
# =========================================
print("\nData Types BEFORE")
print(df.dtypes)

df['temperature'] = df['temperature'].astype(float)
df['humidity'] = df['humidity'].astype(float)
df['energy_usage'] = df['energy_usage'].astype(float)

print("\nData Types AFTER")
print(df.dtypes)


# =========================================
# 4. String Cleaning
# =========================================
print("\nStatus BEFORE Cleaning")
print(df['status'].unique())

df['status'] = df['status'].str.upper().str.strip()

print("\nStatus AFTER Cleaning")
print(df['status'].unique())


# =========================================
# 5. Filtering and Sorting
# =========================================
print("\nDataset BEFORE Sorting")
display(df.head())

df = df.sort_values(by='energy_usage',
                    ascending=False)

print("\nDataset AFTER Sorting")
display(df.head())


# =========================================
# Save Cleaned Dataset
# =========================================
df.to_csv("cleaned_smart_home_dataset.csv",
          index=False)

print("\nCleaned Dataset Saved")
```

Missing Values:
Removed or handled missing sensor values to ensure accurate analysis

Duplicates :
Eliminated duplicate records to prevent biased results

Data Types :
Converted columns (temperature, humidity, energy, timestamp) into proper formats for consistency

Data Standardization :
Standardized categorical values (e.g., ON/OFF status, device types)

Data Integration :
Combined related data into structured tables for better analysis

Sorting & Organization :
Organized data based on timestamp for time-series analysis

Part 3 : DATA PREPARATION & ANALYSIS

```python
# =========================================
# PART 3: DATA PREPARATION & ANALYSIS
# =========================================

import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load Cleaned Dataset
df = pd.read_csv("cleaned_smart_home_dataset.csv")

# =========================================
# 1. Data Summarization
# =========================================
print("Statistical Summary")
print(df[['temperature',
          'humidity',
          'energy_usage']].describe())

print("\nMedian Values")
print(df[['temperature',
          'humidity',
          'energy_usage']].median())


# =========================================
# 2. Group-Based Analysis
# =========================================
group_analysis = df.groupby(
    'device_type'
)['energy_usage'].mean()

print("\nAverage Energy Usage by Device")
print(group_analysis)


# =========================================
# 3. Standardization
# =========================================
scaler = StandardScaler()

df[['scaled_temp',
    'scaled_humidity',
    'scaled_energy']] = scaler.fit_transform(
    df[['temperature',
        'humidity',
        'energy_usage']]
)

print("\nScaled Data")
display(df[['scaled_temp',
            'scaled_humidity',
            'scaled_energy']].head())

print("\nReason:")
print("Scaling places all features on the same scale.")


# =========================================
# 4. Outlier Detection using IQR
# =========================================
Q1 = df['energy_usage'].quantile(0.25)
Q3 = df['energy_usage'].quantile(0.75)

IQR = Q3 - Q1

print("\nDataset Shape BEFORE Outlier Removal")
print(df.shape)

# Remove Outliers
df_clean = df[
    (df['energy_usage'] >= Q1 - 1.5 * IQR) &
    (df['energy_usage'] <= Q3 + 1.5 * IQR)
]

print("\nDataset Shape AFTER Outlier Removal")
print(df_clean.shape)


# =========================================
# Save Final Dataset
# =========================================
df_clean.to_csv("final_prepared_dataset.csv",
                index=False)

print("\nFinal Prepared Dataset Saved")
```

1. Data Summarization

Statistical measures such as mean and median were calculated for temperature, humidity, and energy usage to understand data distribution

Group-based analysis was performed on device types and time to identify usage patterns

2. Standardization

Z-score scaling was applied to normalize numerical features such as temperature, humidity, and energy usage

Reason: Scaling ensures all features are on the same scale, improving analysis and model performance

3. Outlier Detection and Removal

Outliers were detected using the IQR (Interquartile Range) method and removed

Reason: Extreme values in sensor data can distort analysis and lead to incorrect insights

4. Comparison of Data

Dataset size and distribution were compared before and after outlier removal to evaluate the impact of cleaning

PART 4: Visualization using Matplotlib

```python
# =========================================
# PART 4: MATPLOTLIB VISUALIZATION
# =========================================

import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("final_prepared_dataset.csv")


# =========================================
# 1. Line Plot
# =========================================
avg_energy = df.groupby(
    'device_type'
)['energy_usage'].mean()

plt.figure()
avg_energy.plot(kind='line',
                marker='o')

plt.title("Average Energy Usage by Device")
plt.xlabel("Device Type")
plt.ylabel("Average Energy Usage")
plt.grid(True)
plt.show()


# =========================================
# 2. Bar Plot
# =========================================
device_count = df['device_type'].value_counts()

plt.figure()
device_count.plot(kind='bar')

plt.title("Device Usage Count")
plt.xlabel("Device Type")
plt.ylabel("Count")
plt.xticks(rotation=30)
plt.show()


# =========================================
# 3. Scatter Plot
# =========================================
plt.figure()

plt.scatter(df['temperature'],
            df['energy_usage'])

plt.title("Temperature vs Energy Usage")
plt.xlabel("Temperature")
plt.ylabel("Energy Usage")
plt.show()


# =========================================
# 4. Histogram
# =========================================
plt.figure()

plt.hist(df['energy_usage'],
         bins=10)

plt.title("Energy Usage Distribution")
plt.xlabel("Energy Usage")
plt.ylabel("Frequency")
plt.show()
```

**PART 5: Visualization using Seaborn**

```python
# =========================================
# PART 5: SEABORN VISUALIZATION
# =========================================

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("final_prepared_dataset.csv")

# Style
sns.set(style="whitegrid")


# =========================================
# 1. Countplot
# =========================================
plt.figure()

sns.countplot(x='device_type',
              data=df)

plt.title("Count of Device Usage")
plt.xlabel("Device Type")
plt.ylabel("Count")

plt.show()


# =========================================
# 2. Boxplot
# =========================================
plt.figure()

sns.boxplot(x='device_type',
            y='energy_usage',
            data=df)

plt.title("Energy Usage by Device")
plt.xlabel("Device Type")
plt.ylabel("Energy Usage")

plt.show()


# =========================================
# 3. Violin Plot
# =========================================
plt.figure()

sns.violinplot(x='device_type',
               y='energy_usage',
               data=df)

plt.title("Energy Distribution by Device")
plt.xlabel("Device Type")
plt.ylabel("Energy Usage")

plt.show()


# =========================================
# 4. Barplot
# =========================================
plt.figure()

sns.barplot(x='device_type',
            y='energy_usage',
            data=df)

plt.title("Average Energy Usage per Device")
plt.xlabel("Device Type")
plt.ylabel("Average Energy Usage")

plt.show()


# =========================================
# 5. Heatmap
# =========================================
plt.figure()

corr = df[['temperature',
           'humidity',
           'energy_usage']].corr()

sns.heatmap(corr,
            annot=True)

plt.title("Correlation Heatmap")

plt.show()
```
